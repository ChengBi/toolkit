```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([np.random.random(100)-0.5 for i in range(100)])
def gaussian(inputs):
    d = np.array(sorted(inputs))
    return 1./np.sqrt(2*3.14)*np.exp(-0.5*(d*d))
```

```python
data = np.array([gaussian(x[i,:]) for i in range(100)])
for i in range(5):
    plt.plot(data[i,:])
```
